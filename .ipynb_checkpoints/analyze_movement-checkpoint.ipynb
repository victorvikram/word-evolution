```python
avg_yty_change_dim = np.reshape(avg_yty_change, (1, -1))
past_yty_change, futu_yty_change = dist.make_shift_row_arrs(yty_changes)
bavg_yty_change = np.broadcast_to(avg_yty_change_dim, (past_yty_change.shape[0], past_yty_change.shape[1]))
```

```python
print(past_yty_change.shape)
print(futu_yty_change.shape)
print(bavg_yty_change.shape)
```

```python
past_flat = past_yty_change.flatten()
futu_flat = futu_yty_change.flatten()
bavg_flat = bavg_yty_change.flatten()

print(past_flat.shape)
print(futu_flat.shape)
print(bavg_flat.shape)

x_simple = sm.add_constant(bavg_flat.reshape((-1, 1)))
x_detailed = sm.add_constant(np.column_stack((past_flat, bavg_flat)))
print(x_simple.shape)
print(x_detailed.shape)
model_simple = sm.OLS(futu_flat, x_simple).fit()
model_detailed = sm.OLS(futu_flat, x_detailed).fit()

print(model_simple.summary())
print(model_detailed.summary())
```
